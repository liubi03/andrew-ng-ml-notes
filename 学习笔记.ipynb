# 梯度下降算法
梯度下降算法，是机器学习中常用的一种算法，其目是找到参数w和b，使得J(w,b)最小。

梯度下降算法的步骤：
- 初始化参数w和b，一般初始值为0
- 迭代：
    - 计算梯度
    - 沿梯度最速下降方向，更新参数w和b
- 停止：
    - 满足停止条件
最终会到达一个局部最小值，不同初始值，可能会得到不同的结果，陷入局部最优。

可视化：目的是达到某个最低点
<center>



</center>


数学表达：
- 每一步新的参数$w=w-\alpha\frac{\partial J(w,b)}{\partial w}$，其中，等号表示赋值，等号左侧的是新参数，等号右侧的是旧参数。
    - 其中，$\alpha$是学习率，表示学习的步长，一般取很小的正数，学习率过大，可能会导致参数无法收敛，学习率过小，可能会导致参数收敛速度慢。
- 接着更新另一个参数$b=b-\alpha\frac{\partial J(w,b)}{\partial b}$
- 对于梯度下降，是让所有参数同步更新，注意以下的正确和错误示例：

- 注意参数w和b变量的使用，b在更新的时候，使用旧的参数w，而不是更新之后的新的。

- 梯度下降的每一步，使用所有训练样本去计算梯度，而非训练数据的一个子集，然后更新参数w和b。

# 多特征

最初的房价预测的线性回归里面，只有房子的大小一个特征，但是实际的房价预测问题，往往不止一个特征，例如房子的大小，房子的位置，房子的年龄等。

在本节课的例子中，引入到四个特征：
- 分别用$x_1,x_2,x_3,x_4$表示四个特征
- 特征数量用$n$表示
- $\vec{x}^{(i)}=(x_1^{(i)},x_2^{(i)},x_3^{(i)},x_4^{(i)})$表示第$i$个训练样本的特征向量
- $\vec{x}^{(i)}_j$表示第$i$个训练样本的第$j$个特征
- 新的函数f为：$f_{w,b}(x_1,x_2,x_3,x_4)=w_1x_1+w_2x_2+w_3x_3+w_4x_4+b$
    - 更简洁的写法是：$f_{\vec{w},b}(\vec{x})=\vec{w}\dot{\vec{x}}+b$
    - 其中，$\vec{w}$是参数向量，$\vec{x}$是特征向量，$b$是偏置项
    - $\vec{w}\dot\vec{x}$表示参数向量和特征向量的点积

# 向量化

在numpy中，可以用数组来表示向量，例如：
```python
w =np.array([1.0,2.5,-3.3])
b = 4
x =np.array([10,20,30])
```

如果不使用向量化，写代码无疑是低效的：
```python
f = w[0]*x[0] + w[1]*x[1] + w[2]*x[2] + b
```

使用向量化就简洁得多，同时也可以在计算机上并行计算：
- 数学上可以表示为：$f_{\vec{w},b}(\vec{x})=\vec{w}\dot{\vec{x}}+b$
- 实现：
```python
f = np.dot(w,x) + b
```

# 特征缩放
特征缩放，可以使梯度下降更快：
- $x_1$与$x_2$的范围大小不同，可能对最终的f结果影响相同的情况下，$x_1$的变化远大于$x_2$的变化，导致梯度下降的收敛速度慢。
- 此时对特征的进行缩放，将$x_1$和$x_2$都缩放到(0,1)，梯度下降能更快地收敛。
- 如何对特征进行缩放：
    方法一：除以特征的最大值
    - 如$x_1$的范围为[300,2000]，放缩之后$x_{1,scaled}=\frac{x_1}{2000}$，范围为[0.15,1.00]
    方法二：均值归一化
    - 都缩放到(-1,1)之间
    - 公式：$\x_{i,scaled}=\frac{x_i-\mu_i}{x_{i,max}-x_{i,min}}$,其中$\mu_i$是第$i$个特征的均值
    方法三：Z-score归一化
    - 需要计算每个特征的标准差
    - 公式：$\x_{i,scaled}=\frac{x_i-\mu_i}{\sigma_i}$,其中$\mu_i$是第$i$个特征的均值，$\sigma_i$是第$i$个特征的标准差
- 一般特征缩放没有坏处
