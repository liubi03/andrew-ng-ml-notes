## 何为机器学习
- 定义：在进行特定编程的情况下，给予计算机学习能力的领域

机器学习两种主要类型：
- 有监督和无监督学习

## 有监督学习
从输入 x 预测输出 y 的过程，关键特征是训练数据集里面包括正确的数据，机器能够学会在新的x输入下，给出正确的y输出。

两种类型：
- 回归任务：预测一个连续值，如房价预测问题
- 分类任务：预测一个离散值，如肿瘤分类问题

## 无监督学习
无监督学习是指得到的数据中没有任何标签，或者是拥有相同的标签，即标签y，目的是找到数据之间的结构或规律。

- 聚类算法会把数据分成不同的几个簇。

常见的无监督学习算法有：
- 聚类算法常常用于新闻算法，把相似的新闻聚在一起，用于新闻推送。
- 异常检测算法：常用于异常检测，找出数据中的异常点。
- 降维算法：将大数据集压缩成小数据集，尽可能减少信息丢失

## 成本函数 cost function
成本函数，用来评估模型的预测效果。

接着上一节的内容，不同的w和b，会得到不同的函数f，因此会得到不同的预测值。

对这个房价预测模型，我们构建一个成本函数，用来评估预测效果：
- $J(w,b)=\frac{1}{2m}\sum_{i=1}^{m}(\hat{y}^{{i}}-y^{(i)})^2$,即半均方误差，多除以2是为了消去求导后平方产生的系数2，使得计算更加整齐
- 在不同的应用，使用不同的成本函数，半均方误差很常用。我们希望找到w和b，使得J(w,b)最小。

# 梯度下降算法
梯度下降算法，是机器学习中常用的一种算法，其目是找到参数w和b，使得J(w,b)最小。

梯度下降算法的步骤：
- 初始化参数w和b，一般初始值为0
- 迭代：
    - 计算梯度
    - 沿梯度最速下降方向，更新参数w和b
- 停止：
    - 满足停止条件
最终会到达一个局部最小值，不同初始值，可能会得到不同的结果，陷入局部最优。

可视化：目的是达到某个最低点


数学表达：
- 每一步新的参数$w=w-\alpha\frac{\partial J(w,b)}{\partial w}$，其中，等号表示赋值，等号左侧的是新参数，等号右侧的是旧参数。
    - 其中，$\alpha$是学习率，表示学习的步长，一般取很小的正数，学习率过大，可能会导致参数无法收敛，学习率过小，可能会导致参数收敛速度慢。
- 接着更新另一个参数$b=b-\alpha\frac{\partial J(w,b)}{\partial b}$
