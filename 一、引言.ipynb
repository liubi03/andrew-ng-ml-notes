# 1.2 何为机器学习
- Samuel机器学习的定义：在进行特定编程的情况下，给予计算机学习能力的领域
- Tom定义的机器学习：一个程序被认为能从经验E中学习，解决任务T，达到性能度量值P，当且仅当，有了经验E后，经过P评判，程序在处理T时的性能有所提升。
    在垃圾邮件分类任务中：
    - 任务T：将邮件分类为垃圾邮件或非垃圾邮件
    - 性能度量值P：正确分类的邮件比例
    - 经验E：已分类的邮件样本



# 1.3 监督学习
监督学习是我们给学习算法一个数据集。这个数据集由“正确答案”组成。
- 在房价预测例子中，给了一系列房子的数据，我们给定数据集中每个样本的正确价格，即它们实际的售价然后运用学习算法，算出更多的正确答案。

回归问题：预测输出连续的值，例如上面的放假预测.输出一个价格值。
分类问题：预测输出离散的值，例如上面的垃圾邮件分类，输出0或者1。

在其它一些机器学习问题中，通常会遇到不止一种特征
- 例如，我们不仅知道肿瘤的尺寸，还知道对应患者的年龄


# 1.4 无监督学习
无监督学习：数据集没有任何标签，或所有数据标签相同，算法需要自行从数据中发现结构。

聚类：算法自动将数据分成多个不同的“簇”（cluster），簇内数据相似，簇间数据不同。
- 例如，将用户分成不同的“兴趣群”，每个兴趣群的用户都有相似的兴趣。
